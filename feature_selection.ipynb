```python
def cosine_squared(u, v):
    uTv = np.matmul(np.transpose(u),v)
    uTu = np.matmul(np.transpose(u),u)
    vTv = np.matmul(np.transpose(v),v)
    return pow(uTv, 2)/(uTu * vTv)
```
